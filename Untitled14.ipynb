**Data Cleaning with Python and Pandas**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
```

```python
path = "/content/drive/MyDrive/data csv/data.csv"
df = pd.read_csv(path)
print(df)
```

# **Identifying Null Values**

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

# **Detecting Null Values**

```python
sns.heatmap(df.isnull(), annot=True)
```

# **Removing the null values**

```python
df.dropna()
```

# Filling the data by METHOD 1

```python
df.fillna(0)
```

# **Filling the data by METHOD 2**

```python
df.fillna(method = 'ffill')  #FORWARD FILL
```

# **Filling the data by METHOD 3**

```python
df.fillna(method = 'bfill')
```
